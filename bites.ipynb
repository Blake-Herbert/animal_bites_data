```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bites = pd.read_csv('Louisville_Metro_KY_Environmental_Health_Bulk_Data_Animal_Bites_Bulk_Data_3438179908666991130.csv')
bites.head()
```

```python
bites.sort_values(by='bite_date')
```

```python
bites['id'].nunique()
```

There appear to be duplicates of every id, so I'll remove them.

```python
bites.drop_duplicates(subset='id', keep='first', inplace=True)
bites.head()
```

I noticed that there are two breed columns. Many records only have information in one of these columns, however, some mix breed dogs have both columns. 

```python
def combine_breeds(row):
    b1 = row['BreedIDDesc']
    b2 = row['breed']
    if pd.notna(b1) and pd.notna(b2):
        return f'MIX'
    elif pd.notna(b1):
        return b1
    else:
        return b2

bites['breed_cleaned'] = bites.apply(combine_breeds, axis=1)
bites['breed_cleaned'].nunique()
```

```python
bites.sort_values(by='bite_date').head(5)
```

The bite_date column is in date time but the time is 00:00:00 for all entries.

```python
bites['bite_date'] = pd.to_datetime(bites['bite_date']).dt.date
```

```python
bites.sort_values(by='bite_date').head()
```

```python
bites['bite_date'] = pd.to_datetime(bites['bite_date'])
bites['year'] = bites['bite_date'].dt.year
bites_per_year = bites.groupby('year').size()
bites_per_year
```

The current year is 2025, so I'm going to remove records from 2026 and 2027. It also appears that 2020 and earlier have incomplete data, so I'll be removing them as well.

```python
bites = bites[bites['year'] <= 2025]
bites = bites[bites['year'] > 2020]
bites['year'].value_counts().sort_index()
```

```python
bites.isnull().sum() 
```

```python
bites.drop(['vaccination_yrs', 'vaccination_date' ,'AdvIssuedYNDesc', 'quarantine_date', 'DispositionIDDesc', 'head_sent_date', 'release_date', 'ResultsIDDesc'], axis=1, inplace=True)
```

```python
bites.count()
```

```python
counts = bites['SpeciesIDDesc'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(counts.index, counts.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Species')
plt.ylabel('Number of Bites')
plt.title('Bites by Species')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
no_dog_cat = bites[~bites['SpeciesIDDesc'].isin(['DOG', 'CAT'])]

counts = no_dog_cat['SpeciesIDDesc'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(counts.index, counts.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Species')
plt.ylabel('Number of Bites')
plt.title('Bites by Species (excluding Dog and Cat)')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
bites_per_year = bites.groupby('year')['id'].count()

plt.figure(figsize=(10, 6))
plt.bar(bites_per_year.index, bites_per_year.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Year')
plt.ylabel('Number of Bites')
plt.title('Bites per Year')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
years = bites_per_year.index.astype(int)


plt.figure(figsize=(10, 6))
plt.plot(years, 
         bites_per_year.values, 
         marker='o')

plt.ylim(0, 1500)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Year')
plt.ylabel('Number of Bites')
plt.title('Bites per Year')
plt.xticks(years, rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
